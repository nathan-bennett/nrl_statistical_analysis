```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("~/Downloads/nrl.xlsx", skiprows=1)
```

```python
df['margin'] = df['Home Score'] - df['Away Score']
```

```python
home_quantiles = [round(df['Home Odds'].quantile(i), 2) for i in np.arange(0, 1.1, 0.2)]
```

```python
df['home_odds_group'] = np.nan
for index, row in df.iterrows():
    for i in range(len(home_quantiles)):
        if row['Home Odds'] >= home_quantiles[i-1] and row['Home Odds'] < home_quantiles[i]:
            df.loc[index, 'home_odds_group'] = "home odds between: {} - {}".format(home_quantiles[i-1], home_quantiles[i])
        
```

```python
away_quantiles = [round(df['Away Odds'].quantile(i), 2) for i in np.arange(0, 1.1, 0.2)]
```

```python
df['away_odds_group'] = np.nan
for index, row in df.iterrows():
    for i in range(len(away_quantiles)):
        if row['Away Odds'] >= away_quantiles[i-1] and row['Away Odds'] < away_quantiles[i]:
            df.loc[index, 'away_odds_group'] = "away odds between: {} - {}".format(away_quantiles[i-1], away_quantiles[i])
        
```

```python
s = df.groupby(["home_odds_group", "away_odds_group"])['margin'].mean().round(2)
s.unstack()
```

```python
s = df.groupby(["home_odds_group", "away_odds_group"])['margin'].count().round(2)
s.unstack()
```
